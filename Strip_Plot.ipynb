   or

# Orientation:

```python
sns.stripplot(df6.origin, df6.horsepower, palette = 'spring');
```

# jitter:

```python
sns.stripplot(df6.origin, df6.horsepower, jitter = False, palette = 'winter');
```

```python
sns.stripplot(df6.origin, df6.horsepower, jitter = 0.2);
```

```python
from ipywidgets import interactive
def width(j = 0.2):
    sns.stripplot(df6.origin, df6.horsepower,jitter = j)
interactive(width, jitter = (0.1,0.6,0.4))
```

# alpha:

```python
sns.stripplot(x = 'origin', y = 'horsepower', data = df6, 
                                              alpha = 0.7)
```

# size:

```python
sns.stripplot(x = 'origin', y = 'horsepower', data = df6, 
                                      alpha = 0.9, size = 7);
```

# linewidth:

```python
sns.stripplot(x = 'origin', y = 'horsepower', data = df6, 
                                      alpha = 0.9, 
                                      size = 7,
                                      linewidth = 1);
```

# edgecolor:

```python
sns.stripplot(x = 'origin', y = 'horsepower', data = df6, 
                                      alpha = 0.9, 
                                      size = 7,
                                      linewidth = 1,
                                      edgecolor = 'k');
```

# Marker Style:

```python
sns.stripplot(x = 'origin', y = 'horsepower', data = df6, 
                                      alpha = 0.9, 
                                      size = 7,
                                      jitter = 0.2,
                                      linewidth = 1,
                                      edgecolor = 'k',
                                      marker = 'D');
```

# palette:

```python
sns.stripplot(x = 'origin', y = 'horsepower', data = df6, 
                                      alpha = 0.9, 
                                      size = 7,
                                      jitter = 0.2,
                                      linewidth = 1,
                                      edgecolor = 'k',
                                      marker = 'D',
                                      palette = 'plasma');
```

# Multivariate Analysis:

# hue:

```python
sns.stripplot(x = 'origin', y = 'horsepower', data = df6, 
                                      alpha = 0.9, 
                                      size = 7,
                                      jitter = 0.2,
                                      linewidth = 1,
                                      edgecolor = 'k',
                                      marker = 'D',
                                      hue = 'cylinders',
                                      palette = 'Accent'
                                      );
```

# dodge:

```python
sns.stripplot(x = 'origin', y = 'horsepower', data = df6, 
                                      alpha = 0.9, 
                                      size = 7,
                                      jitter = 0.2,
                                      linewidth = 1,
                                      edgecolor = 'k',
                                      marker = 'D',
                                      hue = 'cylinders',
                                      palette = 'PiYG',
                                      dodge = True
                                      );
```

```python
sns.stripplot(x = 'origin', y = 'horsepower', hue = 'model_year', data = df6, )
plt.legend(loc = 1);
```

```python
sns.stripplot(x = 'origin', y = 'horsepower', hue = 'model_year', data = df6,
                                                       dodge = True)
plt.legend(loc = 1);
```

```python
sns.stripplot(x = 'origin', y = 'horsepower', hue = 'cylinders', data = df6,
                                                palette = 'rainbow')
plt.legend(loc = 1)
```

```python
sns.stripplot(x = 'origin', y = 'horsepower', hue = 'cylinders', data = df6,
                                              dodge = True,
                                              palette = 'cubehelix',orient = 'h')
plt.legend(loc = 1);
```
