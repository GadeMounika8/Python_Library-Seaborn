# Build clustermap:

```python
num_col = df9.columns[0:5]
```

```python
num_col
```

```python
sns.clustermap(df9[num_col]);
```

```python
sns.clustermap(df9[num_col], figsize = (6,6));
```

# standard_scale:

```python
sns.clustermap(df9[num_col], standard_scale = 1);
```

# method:

```python
sns.clustermap(df9[num_col], standard_scale = 1, method = 'single');
```

# metric:

```python
sns.clustermap(df9[num_col], standard_scale = 1, method = 'single', metric = 'cityblock');
```
