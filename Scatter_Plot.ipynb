# Bivariate Analysis:

```python
sns.set_style('dark')
```

```python
sns.scatterplot(df7.mpg, df7.horsepower);
```

          or

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', data = df7);
```

           or

```python
sns.scatterplot(x = df7.mpg, y = df7.horsepower);
```

# Marker Styling:

```python
sns.scatterplot(x = df7.mpg, y = df7.horsepower, s = 250, marker = '*');
```

# color, edgecolor:-

```python
sns.scatterplot(x = df7.mpg, y = df7.horsepower, s = 250, marker = '*', color = 'y', edgecolor = 'black');
```

# Semantic Variables:

# hue:

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'origin', data = df7);
```

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'origin', data = df7, palette = 'plasma');
```

# Contineous Variables:

# Palette:

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'model_year', data = df7, 
                                                             palette= 'Accent');
```

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                                                        palette = 'rainbow');
```

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                                                        palette = 'PuBu',
                                                        s = 80);
```

# style:

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                                                        palette = 'nipy_spectral_r',
                                                        s = 80,
                                                        style = 'origin',
                                                        x_jitter = 0.9,
                                                        y_jitter = 0.4);
```

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                                                        s = 70,
                                                        palette= 'Accent',
                                                        style = 'origin',
                                                        x_jitter = 1.2,
                                                        y_jitter = 2);
```

# size:

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                                                        s = 50,
                                                        palette= 'seismic',
                                                        style = 'origin',
                                                        x_jitter = 10,
                                                        y_jitter = 6,
                                                        size = 'cylinders');
```

# sizes:

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                                                        s = 50,
                                                        palette= 'PuOr',
                                                        style = 'origin',
                                                        x_jitter = 10,
                                                        y_jitter = 6,
                                                        size = 'cylinders',
                                                        sizes = [100,70,120,50,70]);
```

# alpha:

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                                                        s = 50,
                                                        palette= 'PiYG',
                                                        style = 'origin',
                                                        x_jitter = 10,
                                                        y_jitter = 6,
                                                        size = 'cylinders',
                                                        sizes = [200, 150, 220,80,30],
                                                        alpha = 0.9);
```

# edgecolor:

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                                                        s = 50,
                                                        palette= 'PiYG',
                                                        style = 'origin',
                                                        x_jitter = 10,
                                                        y_jitter = 6,
                                                        size = 'cylinders',
                                                        sizes = [200, 150, 220, 80, 30],
                                                        alpha = 0.9,
                                                        edgecolor = 'k');
```

# linewidth:

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                                                        s = 50,
                                                        palette= 'PiYG',
                                                        style = 'origin',
                                                        x_jitter = 10,
                                                        y_jitter = 6,
                                                        size = 'cylinders',
                                                        sizes = [200, 150, 220, 80, 30],
                                                        alpha = 0.9,
                                                        edgecolor = 'k',
                                                        linewidth = 0.8);
```

```python
sns.scatterplot(x = 'mpg', y = 'horsepower', hue = 'cylinders', data = df7, 
                             style = 'cylinders',  
                             s = 200, 
                             palette = 'magma_r',);
                
```
