# Univariate analysis:

```python
sns.set_style('white')
```

```python
plt.figure(figsize=(6,5))
sns.swarmplot(df5.bill_depth_mm);
```

# color:

```python
plt.figure(figsize=(6,5))
sns.swarmplot(df5.bill_depth_mm, color = 'purple');
```

# Marker:

```python
sns.swarmplot(df5.flipper_length_mm, color = 'g', marker = 'D')
```

```python
sns.swarmplot(df5.body_mass_g, color = 'indigo', marker = '*');
```

# linewidth:

```python
sns.swarmplot(df5.bill_depth_mm, linewidth = 1);
```

# edgecolor:

```python
sns.swarmplot(df5.bill_depth_mm, linewidth = 1, edgecolor = 'k');
```

# linewidth, edgecolor, marker:

```python
sns.swarmplot(df5.bill_depth_mm, linewidth = 1, edgecolor = 'r', marker = '*');
```

# Bivariate Analysis:

```python
sns.swarmplot(x = df5.species, y = df5.bill_length_mm);
```

# palette:

```python
sns.swarmplot(x = df5.species, y = df5.bill_length_mm, palette = 'summer');
```

```python
sns.swarmplot(x = 'island', y = 'bill_length_mm', data = df5,
                                 palette = 'magma')
```

# Multivariate Analysis:

# hue:

```python
sns.swarmplot(x = 'species', y = 'bill_length_mm',
                               hue = 'sex',
                               data = df5,
                               palette = 'winter')
```

# dodge:

```python
sns.swarmplot(x = 'species', y = 'bill_length_mm',
                               data = df5,
                               hue = 'sex',
                               dodge = True,
                               palette = 'Dark2');
```

```python
sns.swarmplot(x = 'species', y = 'bill_length_mm',
                               hue = 'island',
                               data = df5,
                               palette = 'BuPu')
```

```python
sns.swarmplot(x = 'species', y = 'bill_length_mm',
                               hue = 'island',
                               data = df5,
                               palette = 'viridis')
```

# edgecolor, linewidth:

```python
sns.swarmplot(x = 'species', y = 'bill_length_mm',
                               hue = 'island',
                               data = df5,
                               palette = 'viridis',
                               edgecolor = 'k',
                               linewidth = 0.4)
```

# size:

```python
sns.swarmplot(x = 'species', y = 'bill_length_mm',
                               hue = 'island',
                               data = df5,
                               palette = 'rocket',
                               edgecolor = 'k',
                               linewidth = 0.4,
                               size = 6
                               )
```

# with boxplot:

```python
sns.boxplot(x = 'species', y = 'bill_length_mm', data = df5)
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5);
```

```python
sns.boxplot(x = 'species', y = 'bill_length_mm', data = df5)
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'k');
```

```python
sns.boxplot(x = 'species', y = 'bill_length_mm', data = df5, palette = 'PiYG')
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'k', alpha = 0.8);
```

```python
sns.boxplot(x = 'species', y = 'bill_length_mm', data = df5, palette = 'PiYG', whis = np.inf)
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'k', alpha = 0.8);
```

# with violinplot:

```python
sns.violinplot(x = 'species', y = 'bill_length_mm', data = df5)
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5);
```

```python
sns.violinplot(x = 'species', y = 'bill_length_mm', data = df5)
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'k');
```

```python
sns.violinplot(x = 'species', y = 'bill_length_mm', data = df5, palette = 'BuGn')
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'k', alpha = 0.7);
```

```python
sns.violinplot(x = 'species', y = 'bill_length_mm', data = df5, palette = 'BuGn', scale = 'width')
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'k', alpha = 0.7);
```

```python
sns.violinplot(x = 'species', y = 'bill_length_mm', data = df5, palette = 'BuGn', scale = 'width')
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'k', alpha = 0.7, size = 4);
```

```python
sns.violinplot(x = 'species', y = 'bill_length_mm', data = df5, palette = 'BuGn', scale = 'width', inner = None)
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'k', alpha = 0.7, size = 5);
```

```python
sns.violinplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'cyan', scale = 'width', inner = None)
sns.swarmplot(x = 'species', y = 'bill_length_mm', data = df5, color = 'k', alpha = 0.7, size = 5);
```
