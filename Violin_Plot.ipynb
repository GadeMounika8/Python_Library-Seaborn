# linewidth:

```python
sns.violinplot(x = 'cylinders', y = 'horsepower', data = df4,
                                linewidth = 1);
                   
```

# bandwidth:

```python
sns.violinplot(x = 'cylinders', y = 'horsepower', data = df4,
                                linewidth = 1,
                                bw = 0.3);
```

# Multivariate Analysis:

# palette, hue:

```python
sns.violinplot(x = df4.cylinders, y = df4.horsepower,
                                  hue = df4.origin,
                                  palette = 'viridis');
```

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', data = df4,
                                                   hue = 'origin',
                                                   palette = 'spring');
```

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                     data = df4[df4.origin.isin(['usa', 'japan'])],
                     palette = 'rainbow')
plt.legend(loc=2);
```

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                    data = df4[df4.origin.isin(['usa', 'europe'])],
                    palette = 'plasma')
plt.legend(loc=2);
```

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                 data = df4[df4.origin.isin(['europe', 'japan'])],
                 palette = 'Accent')
plt.legend(loc=2);
```

# split:

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                 data = df4[df4.origin.isin(['europe', 'japan'])], 
                 split = True,
                 palette = 'CMRmap')
plt.legend(loc=2);
```

# inner:

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                 data = df4[df4.origin.isin(['europe', 'japan'])], 
                 split = True,
                 inner = 'quartiles',
                 palette = 'BuPu')
plt.legend(loc=9);
```

# scale:

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                 data = df4[df4.origin.isin(['europe', 'japan'])], 
                 split = True,
                 scale = 'count',
                 palette = 'flare')
plt.legend(loc=9);
```

# scale_hue:

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                 data = df4[df4.origin.isin(['europe', 'japan'])], 
                 split = True,
                 scale = 'count',
                 scale_hue = False,
                 palette = 'Greens')
plt.legend(loc=9);
```

```python
(df4[df4.origin.isin(['europe', 'japan'])].groupby('cylinders').origin.value_counts())
```

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                 data = df4[df4.origin.isin(['europe', 'japan','usa'])],
                 palette = 'PuRd',
                 scale = 'count')
plt.legend(loc=2);
```

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                 data = df4[df4.origin.isin(['europe', 'japan','usa'])], 
                 scale = 'count')
plt.legend(loc=2);
```

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                 data = df4[df4.origin.isin(['europe', 'japan'])], 
                 scale = 'count',
                 scale_hue = False,
                 palette = 'BuGn')
plt.legend(loc=2);
```

```python
sns.violinplot( x = 'cylinders', y = 'horsepower', hue = 'origin',
                 data = df4[df4.origin.isin(['europe', 'japan','usa'])], 
                 scale = 'count',
                 scale_hue = False,
                 linewidth = 1)
plt.legend(loc=2);
```
