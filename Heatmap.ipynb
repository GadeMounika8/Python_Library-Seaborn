```python
sns.heatmap(df8.corr());
```

# Color Palette:

```python
sns.heatmap(df8.corr(), cmap = 'RdPu');
```

# Controlling the colorbar: center:-

```python
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0);
```

# Controling the colorbar: vmin, vmax

```python
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0, vmin = -1, vmax = 1);
```

# Annotations:

```python
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0, vmin = -1, vmax = 1, annot = True);
```

```python
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0, vmin = -1, vmax = 1, annot = True);
```

# fmt:

```python
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0, vmin = -1, vmax = 1, annot = True,
                                                                     fmt = ".1f");
```

# annot_kws:

```python
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0, vmin = -1, vmax = 1, annot = True,
                                                                    fmt = ".2f", annot_kws = {
                                                                    'fontsize':12,
                                                                    'fontweight':'bold',
                                                                    'fontfamily':'Times new roman',
                                                                    'color':'w'
                                                                     });
```

# Styling:

# linewidth:

```python
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0, vmin = -1, vmax = 1, annot = True,
                                                                    fmt = ".2f", annot_kws = {
                                                                    'fontsize':12,
                                                                    'fontweight':'bold',
                                                                    'fontfamily':'Times new roman',
                                                                    'color':'w'
                                                                     },
                                                                     linewidth = 0.2);
```

# linecolor:

```python
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0, vmin = -1, vmax = 1, annot = True,
                                                                    fmt = ".2f", annot_kws = {
                                                                    'fontsize':12,
                                                                    'fontweight':'bold',
                                                                    'fontfamily':'Times new roman',
                                                                    'color':'w'
                                                                     },
                                                                     linewidth = 0.1,
                                                                     linecolor = 'w');
```

# xticklables, yticklables:

```python
labels = ['Bill_Length_mm', 'Bill_Depth_mm', 'Flipper_Length_mm', 'Body_Mass_g']
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0, vmin = -1, vmax = 1, annot = True,
                                                                    fmt = ".2f", annot_kws = {
                                                                    'fontsize':12,
                                                                    'fontweight':'bold',
                                                                    'fontfamily':'Times new roman',
                                                                    'color':'w'
                                                                     },
                                                                     linewidth = 0.1,
                                                                     linecolor = 'w',
                                                                     xticklabels = labels, 
                                                                     yticklabels = labels);
```

```python
labels = ['Bill_Length_mm', 'Bill_Depth_mm', 'Flipper_Length_mm', 'Body_Mass_g']
sns.heatmap(df8.corr(), cmap = 'RdPu',center = 0, vmin = -1, vmax = 1, annot = True,
                                                                    fmt = ".2f", annot_kws = {
                                                                    'fontsize':12,
                                                                    'fontweight':'bold',
                                                                    'fontfamily':'Times new roman',
                                                                    'color':'w'
                                                                     },
                                                                     linewidth = 0.1,
                                                                     linecolor = 'w',
                                                                     xticklabels = False, 
                                                                     yticklabels = labels);
```
