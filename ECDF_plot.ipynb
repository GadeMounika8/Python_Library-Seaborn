# Additional Options:

# color & count:

```python
sns.ecdfplot(x = 'total_bill', data = df1,
                               stat = 'count',
                               color = 'indigo')
```

```python
sns.ecdfplot(x = 'total_bill', data = df1,
                               stat = 'count',
                               color = 'gold')
```

# Ranking with count and complementary:

```python
sns.ecdfplot(x = 'total_bill', data = df1,
                               stat = 'count',
                               complementary = True,
                               color = 'black');
```

# weights:

```python
sns.ecdfplot(x= 'total_bill', data = df1, 
                     weights = 'total_bill', 
                     color = 'olive');
plt.axvline(0.4, c = 'red');
```

# Styling:

# color:

```python
sns.ecdfplot(x = 'tip', data = df1, 
                  color = 'cyan');
```

```python
sns.ecdfplot(x = 'size', data = df1, 
                    color = 'purple');
```

# Multivariate Analysis:

```python
sns.ecdfplot(x = 'total_bill', data = df1,
                               hue = "sex");
```

# palette:

```python
sns.ecdfplot(x = 'total_bill', data = df1,
                               hue = "time",
                               palette = 'BuGn');
```

```python
sns.ecdfplot(x = 'total_bill', data = df1,
                               hue = "day",
                               palette = 'twilight');
```

```python
sns.ecdfplot(x = 'total_bill', data = df1,
                               hue = "smoker",
                               palette = 'PuBu');
```

```python
sns.ecdfplot(x = 'total_bill', data = df1,
                               hue = 'day',
                               palette = 'ocean');
```

# linewidth:

```python
x = sns.ecdfplot(x = 'total_bill', data = df1,
                               hue = 'day',
                               palette = 'CMRmap',
                               lw = 1);
```

```python
type(x)
```
