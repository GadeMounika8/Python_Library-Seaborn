# Styling:

# color, kdeplot:

```python
sns.set_style('dark')
sns.histplot(x = 'total_bill', data = df,
                            kde = True, color = 'y');
```

# bins:

```python
sns.set_style('dark')
sns.histplot(x = 'total_bill', data = df,
                            kde = False,
                            bins = 15, color = 'indigo');
```

```python
sns.set_style('dark')
sns.histplot(x = 'total_bill', data = df,
                            kde = False,
                            bins = [3,10,25,35,50], color = 'olive');
```

# bindwidth:

```python
sns.set_style('dark')
sns.histplot(x = 'total_bill', data = df,
                            kde = False,
                            binwidth = 5, color = 'cyan');
```

# binrange:

```python
sns.set_style('dark')
sns.histplot(x = 'total_bill', data = df,
                            kde = True,
                            binwidth = 6,
                            binrange = (0,60), color = 'orange');
```

# Statistics:

# stat:

```python
sns.set_style('dark')
sns.histplot(x = 'total_bill', data = df,
                               stat = 'count', color = 'green');
```

```python
sns.set_style('dark')
sns.histplot(x = 'total_bill', data = df,
                               stat = 'density', color = 'purple');
```

# cumulative:

```python
sns.set_style('white')
sns.histplot(x = 'total_bill', data = df,
                               stat = 'probability',
                               element = 'step',
                               cumulative = False, color = 'gold');
```

```python
sns.set_style('dark')
sns.histplot(x = 'total_bill', data = df,
                               stat = 'probability',
                               fill = False,
                               element = 'step',
                               cumulative = False, color = 'k');
```

```python
sns.set_style('white')
sns.histplot(x = 'total_bill', data = df,
                               stat = 'probability',
                               fill = True,
                               element = 'step',
                               cumulative = True, color = 'violet');
```
