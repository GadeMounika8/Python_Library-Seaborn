# Estimator Parameter:

```python
sns.barplot(x = 'day', y = 'tip', data= df10,
                                     ci = 95, 
                                     estimator = np.min,
                                     palette = 'ocean');
```

```python
sns.barplot(x = 'day', y = 'tip', data= df10,
                                     ci = 95, 
                                     estimator = np.max,
                                     palette = 'flare');
```

```python
sns.barplot(x = 'day', y = 'tip', data= df10,
                                     ci = 95, 
                                     estimator = np.std,
                                     palette = 'viridis');
```

```python
sns.barplot(x = 'day', y = 'tip', data= df10,
                                     ci = 95, 
                                     estimator = np.mean,
                                     palette = 'Dark2');
```

```python
sns.barplot(x = 'day', y = 'tip', data= df10,
                                 ci = 95, 
                                 estimator = np.median,
                                 palette = 'plasma'
                                 );
```

# dodge:

# capsize:

```python
sns.barplot(x = 'day', y = 'tip', data= df10, 
                                  hue = 'sex',
                                  order = ['Sun', 'Sat', 'Fri', 'Thur'],
                                  hue_order = ['Female', 'Male'],
                                  errcolor = 'k',
                                  errwidth = 1.6,
                                  palette = 'rainbow_r',
                                  edgecolor = 'k',
                                  linewidth = '1.2',
                                  dodge = True,
                                  capsize =0.1);
```
